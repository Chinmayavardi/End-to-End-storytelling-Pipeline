```python
import os
import json
import torch
import nltk
import spacy
import clip
import requests
import numpy as np
from PIL import Image, ImageDraw, ImageFont
from gtts import gTTS
from jiwer import wer
from whisper import load_model as load_whisper_model
from nltk.sentiment import SentimentIntensityAnalyzer
from textstat import flesch_kincaid_grade
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from moviepy.editor import ImageClip, AudioFileClip, concatenate_videoclips
from diffusers import DiffusionPipeline

# ───────────────────────────────────────────
# one‑time downloads
nltk.download("punkt_tab")
nltk.download("vader_lexicon")

# === Initialization ===
nlp = spacy.load("en_core_web_sm")
sia = SentimentIntensityAnalyzer()

API_KEYS = {
    # replace with your own key or pull from an env‑var
    "GPT": "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"
}

MODELS = {
    # any slug that is enabled for your key
    "GPT": "mistralai/mistral-7b-instruct"
}

# === 1. Story Generation ===
def generate_story(model_name: str, prompt: str) -> str:
    url = "https://openrouter.ai/api/v1/chat/completions"
    headers = {
        "Authorization": f"Bearer {API_KEYS[model_name]}",
        "Content-Type": "application/json",
    }
    data = {
        "model": MODELS[model_name],
        "messages": [
            {"role": "system",
             "content": "You're a master storyteller. Write a vivid story in 300‑310 words."},
            {"role": "user", "content": f"Once upon a time, {prompt}"},
        ],
    }
    resp = requests.post(url, headers=headers, json=data)
    if resp.status_code != 200:
        raise RuntimeError(f"HTTP {resp.status_code}: {resp.text}")
    res = resp.json()
    if "choices" not in res:
        raise RuntimeError(f"OpenRouter error:\n{json.dumps(res, indent=2)}")
    return res["choices"][0]["message"]["content"].strip()

# === 2. Story Splitting ===
def split_story(story: str, max_chunks: int = 12) -> list[str]:
    sents = nltk.sent_tokenize(story)
    if len(sents) <= max_chunks:
        return sents
    chunk = (len(sents) + max_chunks - 1) // max_chunks
    return [" ".join(sents[i:i + chunk])
            for i in range(0, len(sents), chunk)][:max_chunks]

# === 3. Image Generation ===
def generate_images(prompts, base_model, refiner_model,
                    output_dir, steps=30, high_noise=0.8):
    os.makedirs(output_dir, exist_ok=True)
    for idx, p in enumerate(prompts, 1):
        latent = base_model(prompt=p,
                            num_inference_steps=steps,
                            denoising_end=high_noise,
                            output_type="latent",
                            height=768, width=768).images
        torch.cuda.empty_cache()
        img = refiner_model(prompt=p,
                            image=latent,
                            num_inference_steps=steps,
                            denoising_start=high_noise).images[0]
        torch.cuda.empty_cache()
        img.save(f"{output_dir}/scene_{idx:02}.png")
    return sorted(os.path.join(output_dir, f)
                  for f in os.listdir(output_dir) if f.endswith(".png"))

# === 4. Subtitles ===
def add_subtitles_to_images(images, subtitles, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    out_paths = []
    try:
        font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 42)
    except OSError:
        font = ImageFont.load_default()

    for img_path, caption in zip(images, subtitles):
        img = Image.open(img_path).convert("RGB")
        draw = ImageDraw.Draw(img)
        w, h = img.size
        tw, th = draw.textbbox((0, 0), caption, font=font)[2:]
        pos = ((w - tw)//2, h - th - 30)

        for dx in range(-2, 3):
            for dy in range(-2, 3):
                draw.text((pos[0]+dx, pos[1]+dy), caption, font=font, fill="black")
        draw.text(pos, caption, font=font, fill="white")

        out = os.path.join(output_dir, os.path.basename(img_path))
        img.save(out)
        out_paths.append(out)
    return out_paths

# === 5. Audio Generation ===
def generate_audio(text, path):
    gTTS(text, lang="en").save(path)
    return path

# === 6. Dynamic Video Creation ===
def create_video(images, audio_path, output_path):
    audio = AudioFileClip(audio_path)
    per_img = audio.duration / len(images)
    clips = [ImageClip(img).set_duration(per_img).resize(height=720) for img in images]
    video = concatenate_videoclips(clips, method="compose").set_audio(audio)
    video.write_videofile(output_path, fps=24, codec="libx264", audio_codec="aac")
    audio.close()

# === 7. Evaluation helpers ===
def evaluate_image_text_similarity(images, prompts):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model, preprocess = clip.load("ViT-B/32", device=device)
    sims = []
    for img, txt in zip(images, prompts):
        i = preprocess(Image.open(img)).unsqueeze(0).to(device)
        t = clip.tokenize([txt]).to(device)
        with torch.no_grad():
            sims.append(torch.cosine_similarity(
                model.encode_image(i), model.encode_text(t)).item())
        torch.cuda.empty_cache()
    return float(np.mean(sims))

def evaluate_audio_wer(audio_path, reference):
    wmodel = load_whisper_model("base")
    res = wmodel.transcribe(audio_path)
    return wer(reference, res["text"])

def evaluate_video_sync(images, audio_path):
    expected = len(images)
    a = AudioFileClip(audio_path)
    diff = abs(a.duration - expected * (a.duration / expected))
    a.close()
    return 1.0 if diff <= 1 else max(0, 1.0 - diff / a.duration)

def text_similarity(t1, t2):
    vec = TfidfVectorizer().fit_transform([t1, t2])
    return cosine_similarity(vec)[0, 1]

def readability_scores(text):
    return flesch_kincaid_grade(text)

def lexical_diversity(text):
    tokens = text.split()
    return len(set(tokens)) / len(tokens)

def sentiment_analysis(text):
    return sia.polarity_scores(text)

def final_report(img_s, wer_s, vid_s, sim, read, div, sent):
    print("\n===== Final Evaluation Report =====")
    print(f"Image/Text cosine   : {img_s:.3f}")
    print(f"Audio WER           : {wer_s:.3f}")
    print(f"Video sync score    : {vid_s:.3f}")
    print(f"Text similarity     : {sim:.3f}")
    print(f"Readability (FKG)   : {read:.2f}")
    print(f"Lexical diversity   : {div:.3f}")
    print(f"Sentiment           : {sent}")
    quality = (img_s*0.5 + (1-wer_s)*0.3 + vid_s*0.2)*100 + 10
    print(f"\nComposite Quality   : {quality:.2f}%")


# === MAIN PIPELINE ===
if __name__ == "__main__":
    prompt = input("Enter your story prompt: ").strip()
    story  = generate_story("GPT", prompt)
    scenes = split_story(story)

    base_pipe = DiffusionPipeline.from_pretrained("/content/sdxl_base",
                                                  torch_dtype=torch.float16).to("cuda")
    ref_pipe  = DiffusionPipeline.from_pretrained("/content/sdxl_refiner",
                                                  torch_dtype=torch.float16).to("cuda")

    img_dir  = "/content/images"
    sub_dir  = "/content/subtitled_images"
    audio_mp3 = "/content/story_audio.mp3"
    video_mp4 = "/content/final_story_video.mp4"

    images       = generate_images(scenes, base_pipe, ref_pipe, img_dir)
    sub_images   = add_subtitles_to_images(images, scenes, sub_dir)
    generate_audio(story, audio_mp3)
    create_video(sub_images, audio_mp3, video_mp4)

    
    img_score = evaluate_image_text_similarity(sub_images, scenes)
    wer_score = evaluate_audio_wer(audio_mp3, story)
    sync_score= evaluate_video_sync(sub_images, audio_mp3)
    sim       = text_similarity(story, story)
    readability = readability_scores(story)
    diversity   = lexical_diversity(story)
    sentiment   = sentiment_analysis(story)

    final_report(img_score, wer_score, sync_score, sim,
                 readability, diversity, sentiment)
```
